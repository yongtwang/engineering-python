### Jupyter Notebook

- Type in `cmd` in the Windows Start Menu and hit `Enter` to run the Command Prompt. In Mac OS, you should type in `terminal` in Launchpad. 
<img src="images/command_prompt_1.png" alt="Command Prompt" style="width: 300px;"/>
- Type in `cd Downloads` and hit `Enter` to change directory to the `Downloads` folder. 
- Then type in `cd engineering-python-master` and hit `Enter` to further change directory to the base folder `engineering-python-master`.
- Type in `jupyter notebook` in the Command Prompt or Terminal and hit `Enter`.
<img src="images/command_prompt_2.png" alt="Command Prompt" style="width: 500px;"/>

- Wait for a few seconds until Jupyter Notebook opens in a browser. This is the Dashboard view of Jupyter Notebook.
<img src="images/jupyter_1.png" alt="Jupyter Notebook" style="width: 500px;"/>
- Click and go to the `02C_Jupyter_Notebook` folder. You will see `02C_Jupyter_Notebook.ipynb` and `hello.py`.
<img src="images/jupyter_2.png" alt="Jupyter Notebook" style="width: 500px;"/>
- Click to open `02C_Jupyter_Notebook.ipynb`.
<img src="images/jupyter_3.png" alt="Jupyter Notebook" style="width: 550px;"/>

- Now scroll down and type in a few Python statements below and click the run cell button in the toolbar: <img src="images/run_cell.png" alt="Run Cell Button" style="width: 25px;"/> You will immediately see the results.

```python
print("Hello World")
```

```python
1 + 2
```

- Click to open `hello.py`. You will see the source code in an editor. You can make changes and save the new file.
<img src="images/jupyter_4.png" alt="Jupyter Notebook" style="width: 550px;"/>
