# SymPy Algebra and Calculus

### Limits

$$\lim_{x\to x_0} f(x)$$

```python
from sympy import *
x, y, z = symbols('x, y, z')
init_printing(pretty_print=True)
```

```python
limit(sin(x)/x, x, 0)
```

```python
limit(x, x, oo)                      
```

```python
limit(1/x, x, oo)
```

```python
limit(x**x, x, 0)
```

### Limits from a Certain Direction

$$\lim_{x\to x_0^+} f(x),\hspace{1cm}\lim_{x\to x_0^-} f(x)$$

```python
limit(1/x, x, 0, dir='+')
```

```python
limit(1/x, x, 0, dir='-')
```

### Differentiation (Derivatives)

$$ f'(x),\hspace{1cm}\frac{df}{dx}$$

```python
diff(sin(x), x)
```

```python
diff(sin(2*x), x)
```

### Higher Order Derivatives

```python
diff(sin(2*x), x, 1)
```

```python
diff(sin(2*x), x, 2)
```

```python
diff(sin(2*x), x, 3)
```

### Indefinite Integrals

$$\int_{-\infty}^{\infty} f(x) dx$$

```python
integrate(6 * x**5, x)
```

```python
integrate(sin(x), x)
```

```python
integrate(log(x), x)  # By default the base is e
```

```python
integrate(2*x + sinh(x), x)
```

### Definite Integrals

$$\int_{a}^{b} f(x) dx$$

```python
integrate(x**3, (x, -1, 1))
```

```python
integrate(sin(x), (x, 0, pi/2))
```

```python
integrate(cos(x), (x, -pi/2, pi/2))
```

```python
integrate(exp(-x), (x, 0, oo))
```

```python
integrate(exp(-x**2), (x, -oo, oo))
```

### Taylor Series Expansion of an Expression at a Point

```python
series(ln(x), x, 1, 4)  # At point 1, order 4
```

```python
series(cos(x), x)  # By default, point=0, order=6
```

### Equation Solving

```python
solve(x**2-3*x+2, x)
```

```python
solve([x+5*y-2, -3*x+6*y-15], [x, y])
```

```python
solve(exp(x) + 1, x)
```

```python
f = x**4 - 3*x**2 + 1
f
```

```python
factor(f)
```

```python
f = symbols('f', cls=Function)
f(x)
```

```python
f(x).diff(x)
```

```python
f(x).diff(x, x)
```

```python
f(x).diff(x, x) + f(x)  # Ordinary Differential Equation (ODE)
```

```python
dsolve(f(x).diff(x, x) + f(x), f(x))  # Solving (some) ODEs
```

### Substitution

```python
expr = x**2 + 2*x + 1
expr
```

```python
expr.subs(x, 2)
```

```python
expr.subs(x, y+1)
```

```python
n, k = symbols('n, k', integer=True)
a = factorial(n)
a
```

```python
a.subs(n, 3)
```

```python
b = binomial(n, k)
b
```

```python
b.subs([(n,3), (k,2)])
```

```python
expr = x**3 + 4*x*y - z
expr
```

```python
expr.subs([(x, 2), (y, 4), (z, 0)])
```

### Converting Strings to SymPy Expressions

```python
str_expr = "x**2 + 3*x - 1/2"
expr = sympify(str_expr)  # Not to be confused with simplify
expr
```

```python
expr = S(str_expr)  # Equivalent
expr
```
