### Housing Market

- Quandl is a search engine for financial, economic, and social datasets.
- Download ["Freddie Mac House Price Index NY"](https://www.quandl.com/data/FMAC/HPI_NY) as an Excel file `FMAC-HPI_NY.xls`.
- The Freddie Mac House Price Index (FMHPI) provides a measure of typical price inflation for houses within the United States.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('FMAC-HPI_NY.xls', 'Worksheet1', index_col = 'Date', na_values='NaN')
df
```

```python
df = df.sort_index(axis=0, ascending=True)
df
```

```python
df.plot()
plt.show()
```

### Retrieving Data from Quandl

- The data retrieval process can be done automatically.
- Install the quandl package by executing the following command in a command window or terminal (not within the Python environment): `conda install quandl`
<img src="images/install_quandl.png" alt="Install Quandl" style="width: 550px;"/>

```python
import quandl
```

```python
df = quandl.get('FMAC/HPI_NY')
df
```

- Retrieve the data of four states (CA, IL, NY, TX). This step needs an api_key.
- To apply for a free api_key, you need to log into you quandl account. 
- In your "Account Settings", you will find your api key. You can also request a new api key.
<img src="images/api_key.png" alt="Quandl API Key" style="width: 450px;"/>

```python
api_key = 'PXmEQjwkXi9Cgm2xU8Xs'  # Replace this with your own api key
states = ['CA', 'IL', 'NY', 'TX']  
df = pd.DataFrame  # Create an empty DataFrame

for abbr in states:
    data_label = 'FMAC/HPI_' + abbr
    print(data_label)
    temp_df = quandl.get(data_label, authtoken=api_key)
    if df.empty:
        df = temp_df
    else:
        df = pd.concat([df, temp_df], axis=1)  # Add new columns to df
df.columns = states
print(df)
df.to_excel('FMAC-HPI_CA_IL_NY_TX.xlsx', sheet_name='Sheet1')
```

```python
df.plot()
plt.legend(loc='best')
plt.show()
```

- December 2000 is the base month in FMHPI. 
- We can move the reference point to the beginning, i.e., January 1975.

```python
new_df = (df - df.iloc[0,:]) / df.iloc[0,:]
new_df.plot()
plt.legend(loc='best')
plt.show()
```

```python
ctable = new_df.corr()  # Correlation table
ctable
```
